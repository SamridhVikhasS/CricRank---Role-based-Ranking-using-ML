## Importing required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

## Data Cleaning & Preprocessing

```python
bat = pd.read_csv("Batting Records.csv")
bat
```

```python
bat.drop(bat[bat['innings']=="-"].index,inplace=True)
bat.drop(bat[bat['average']=="-"].index,inplace=True)
bat.drop(bat[bat['strike_rate']=="-"].index,inplace=True)
bat
```

```python
bat.info()
```

```python
bat['innings'] = bat['innings'].astype('int64')
bat['not_out'] = bat['not_out'].astype('int64')
bat['runs'] = bat['runs'].astype('int64')
bat['average'] = bat['average'].astype('float64')
bat['balls_faced'] = bat['balls_faced'].astype('int64')
bat['strike_rate'] = bat['strike_rate'].astype('float64')
bat['100s'] = bat['100s'].astype('int64')
bat['50s'] = bat['50s'].astype('int64')
bat['duck'] = bat['duck'].astype('int64')
bat['4s'] = bat['4s'].astype('int64')
bat['6s'] = bat['6s'].astype('int64')
```

```python
bat.info()
```

```python
bat.duplicated().sum()
```

```python
bat.dropna(axis=0,inplace=True)
```

```python
boundaries = bat["4s"] + bat["6s"]
bat['boundary_per_ball'] = np.round(boundaries/bat['balls_faced'],3)
bat.head()
```

```python
sns.boxplot(bat.boundary_per_ball)
plt.title("Boundaries per Ball")
```

```python
bat['boundary_index'] = np.round(boundaries/bat.innings,3)
bat.head()
```

```python
sns.boxplot(bat.boundary_index)
plt.title("Boundaries index")
```

```python
bat['not_out_pct'] = np.round(bat['not_out']/bat.innings,3)
bat.head()
```

```python
sns.boxplot(bat.not_out_pct)
plt.title("Not out%")
```

```python
bat['duck_index'] = np.round(bat['duck']/bat.innings,3)
bat.head()
```

```python
sns.boxplot(bat.duck_index)
plt.title("Duck index")
```

```python
rwob = bat.runs - 4*bat['4s'] - 6*bat['6s']
bat['RwoB_index'] = np.round(rwob/bat.innings,3)
bat.head()
```

```python
sns.boxplot(bat.RwoB_index)
plt.title("Runs without Boundaries Index")
```

```python
bat['big_match_index'] = np.round(((2 * bat['100s']) + bat['50s']) / bat.innings,3)
bat.head()
```

```python
sns.boxplot(bat.big_match_index)
plt.title("Big Match Index")
```

## Clustering Dataset

```python
batter = bat[['name','country','average','strike_rate','boundary_per_ball','boundary_index','not_out_pct','duck_index','RwoB_index','big_match_index']]
batter.head()
```

```python
batter.describe()
```

```python
cluster_bat = batter.copy().drop(['name','country'],axis=1)
cluster_bat
```

## Preliminary Rank Metrics

```python
bat['prelim'] = np.round((bat.average*bat.strike_rate)/100,3)
bat.head()
```

```python
bat.describe()
```

```python
sns.boxplot(bat.prelim)
```

Ranking using Prelim Metric

```python
bat.loc[((bat.prelim>30) & (bat.runs>=500)),'prelim_rank'] = "Best"
bat.loc[((bat.prelim>30) & (bat.runs<500)),'prelim_rank'] = "Good"
bat.loc[((bat.prelim>20) & (bat.prelim<=30)),'prelim_rank'] = "Good"
bat.loc[((bat.prelim>=10) & (bat.prelim<=20)),'prelim_rank'] = "Average"
bat.loc[((bat.prelim<10)),'prelim_rank'] = "Poor"
bat.loc[((bat.runs<100)),'prelim_rank'] = "Poor"
```

```python
bat
```

```python
sns.countplot(bat.prelim_rank)
```

```python
bat.prelim_rank.value_counts()
```

```python
batter["prelim_metric"] = bat["prelim"]
batter["prelim_rank"] = bat["prelim_rank"]
batter.head()
```

```python
bat_test = batter.sample(frac=0.07,random_state=25)
bat_test.shape
```

```python
bat_train = batter.drop(bat_test.index)
bat_train.shape
```

```python
batter
```

```python
batter.to_csv("Batting Cluster Dataset.csv",index=False)
bat_train.to_csv("Training Batters.csv",index=False)
bat_test.to_csv("Test Batters.csv",index=False)
```

# Feature Scaling

```python
from sklearn import metrics
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
```

```python
col = cluster_bat.columns.tolist()
col
```

#### Normal Data

```python
plt.figure(figsize=(20,10))
sns.boxplot(data = cluster_bat)
plt.show()
```

```python
cluster_bat.describe()
```

```python
fig = plt.figure(figsize=(25,15))
ax = fig.gca()
cluster_bat.hist(ax = ax)
plt.show()
```

```python
sns.heatmap(cluster_bat.corr(),annot=True)
```

```python
wcss = []
for i in range(2,11):
    model = KMeans(i).fit(cluster_bat.dropna())
    wcss.append(model.inertia_)
plt.plot(np.arange(2,11,1),wcss)
```

```python
for cl in range(2,11):
    cluster = KMeans(n_clusters = cl,random_state=10)
    cluster_labels = cluster.fit_predict(cluster_bat)
    
    avg_sil = silhouette_score(cluster_bat,cluster_labels)
    print("For n_clusters = {}, The Average Silhouette Score is {}".format(cl,avg_sil))
```
