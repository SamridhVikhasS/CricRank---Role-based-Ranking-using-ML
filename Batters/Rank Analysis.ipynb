```python
import pandas as pd
from scipy.stats import spearmanr,kendalltau
```

```python
df = pd.read_csv("ICC_allTime_top_10_batter.csv")
df
```

```python
ranks = df[["name","ICC_rank","prelim_rank","RBML_rank","PCA_rank","hussey_rank"]]
ranks
```

```python
ranks.sort_values(by="hussey_rank")[['name','hussey_rank']]
```

#### Prelim Rank Correlation

```python
print(spearmanr(ranks["ICC_rank"],ranks["prelim_rank"]))
print(kendalltau(ranks["ICC_rank"],ranks["prelim_rank"]))
```

#### RBML Rank Correlation

```python
print(spearmanr(ranks["ICC_rank"],ranks["RBML_rank"]))
print(kendalltau(ranks["ICC_rank"],ranks["RBML_rank"]))
```

#### PCA Rank Correlation

```python
print(spearmanr(ranks["ICC_rank"],ranks["PCA_rank"]))
print(kendalltau(ranks["ICC_rank"],ranks["PCA_rank"]))
```

#### Hussey Rank Correlation

```python
print(spearmanr(ranks["ICC_rank"],ranks["hussey_rank"]))
print(kendalltau(ranks["ICC_rank"],ranks["hussey_rank"]))
```
